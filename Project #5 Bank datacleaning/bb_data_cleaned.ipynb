```python
import pandas as pd
from datetime import date
import numpy as np
```

```python
bb = pd.read_csv('bb_data_unclean.csv')
```

```python
bb.head(40)
```

```python
bb= bb.drop(1)
```

```python
bb = bb.drop('0',axis=1)
```

```python
bb.columns =['Serial no.', 'Description', 'Amount in December, 2015' ,'Amount in January, 2016', '% Change (December, 2015 to January, 2016)','year', 'mo']
```

```python
bb = bb.rename(columns={'Amount in January, 2016': 'Target'})
```

```python
bb.head(40)
```

```python
Date = []
for y, m in zip(bb.year, bb.mo):
    Date.append(date(y, m, 1))


bb['Date'] = Date
```

```python
df = bb
```

```python
df.head(40)
```

```python
df=df.drop(columns=['Amount in December, 2015', "% Change (December, 2015 to January, 2016)", 'year','mo'])
```

```python
df=df.drop(columns=['Serial no.'])
```

```python
df2 = df
```

```python
df2.Description.unique()
```

```python
df2= df2[~df2.Description.str.startswith('Mobile Financial Services')]
df2.Description.unique()
```
