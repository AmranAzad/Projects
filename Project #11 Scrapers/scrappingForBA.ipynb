```python
def get_listing_data(url):
    r = requests.get(url)
    soup = bs(r.content, 'html.parser')

    lists = soup.find_all('div', {'class':"word-break--2nyVq value--1lKHt"})
    
    try:
        Price = soup.find("div", class_="amount--3NTpl").text
    except IndexError:
        Price = 'null'

    for list in lists:
        try:
            Brand = lists[0].text
        except IndexError:
            Brand = 'null'
        try:
            Model = lists[1].text
        except IndexError:
            Model = 'null'
        try:
            Edition = lists[2].text
        except IndexError:
            Edition = 'null'
        try:
            Year_of_Manufacture = lists[3].text
        except IndexError:
            Year_of_Manufacture = 'null'
        try:
            Registration_year = lists[4].text
        except IndexError:
            Registration_year = 'null'  
        try:
            Condition = lists[5].text
        except IndexError:
            Condition = 'null' 
        try:
            Transmission = lists[6].text
        except IndexError:
            Transmission = 'null' 
        try:
            Body_type = lists[7].text
        except IndexError:
            Body_type = 'null' 
        try:
            Fuel_type = lists[8].text
        except IndexError:
            Fuel_type = 'null' 
        try:
            Engine_capacity = lists[9].text
        except IndexError:
            Engine_capacity = 'null' 
        try:
            Mileage = lists[10].text
        except IndexError:
            Mileage = 'null' 
    listing_info = {}
    for variable in [
        "Brand", 
        "Model",
        "Edition", 
        "Year_of_Manufacture",
        "Registration_year", 
        "Condition", 
        "Transmission",
        "Body_type",
        "Fuel_type",
        "Engine_capacity", 
        "Mileage",
        "Price"]:
        listing_info[variable]=eval(variable)

    return listing_info
get_listing_data("https://bikroy.com/en/ad/toyota-premio-g-superior-fullfresh-2011-for-sale-dhaka")
```
